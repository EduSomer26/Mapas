```julia
using DynamicalSystems, CairoMakie

ds = Systems.henon() 
i = 1 
p_values = 1:0.0005:1.4
n = 2000
Trt = 2000
p_index = 1

output = orbitdiagram(ds, i, p_index, p_values; n = n, Ttr = Trt)

L = length(p_values)
x = Vector{Float64}(undef, n*L)
y = copy(x)

for j in 1:L
    x[(1 + (j-1)*n):j*n] .= p_values[j]
    y[(1 + (j-1)*n):j*n] .= output[j]
end

fig = Figure(resolution = (500,500))
fig, ax = scatter(x, y; axis = (xlabel = L"a", ylabel = L"x"),
    markersize = 0.8, color = ("black", 0.5),
)
xlims!(ax, p_values[1], p_values[end]); ylims!(ax,-1.5,1.5)
fig
```

```julia
using DynamicalSystems, CairoMakie

ds = Systems.henon(a = 1.4, b = 0.3)

n = 10000
points = trajectory(ds, n; Ttr = 1000)

x, y = columns(points)

fig = Figure(resolution = (500,500))
fig, ax, obj = scatter(x, y; axis = (xlabel = L"x", ylabel = L"y"),
    markersize = 1, color=("black", 1)
)
fig
```
